```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# 1️⃣ Load inputs & predictions
inputs = pd.read_csv("../data/new_input.csv")
preds = pd.read_csv("../predictions/new_predictions.csv")
```

```python
# Combine them
df_results = inputs.copy()
df_results["Predicted Recycling Rate (%)"] = preds["Prediction"]
```

```python
# Preview the merged dataframe
display(df_results.head())
```

```python
# 2️⃣ Bar Chart of Predictions
plt.figure(figsize=(8,5))
plt.bar(range(len(df_results)), df_results["Predicted Recycling Rate (%)"])
plt.xticks(range(len(df_results)), [f"Row {i+1}" for i in range(len(df_results))])
plt.ylabel("Predicted Recycling Rate (%)")
plt.title("Predicted Recycling Rates for New Inputs")
plt.show()
```

```python
# 3️⃣ Map Visualization (requires Latitude & Longitude columns)
fig = px.scatter_mapbox(
    df_results,
    lat="Latitude",
    lon="Longitude",
    size="Predicted Recycling Rate (%)",
    color="Predicted Recycling Rate (%)",
    hover_name="Predicted Recycling Rate (%)",
    zoom=4,
    mapbox_style="open-street-map"
)
fig.show()
```
